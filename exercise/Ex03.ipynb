# 解答案
何をやっているのかひとつずつ追えるようにやや冗長に書いてます。  
とりあえず今は動けばヨシ！

```python
#10進数で表記された緯度経度を60進数に変換して表示する

#東京駅の緯度経度
latlon = [35.68131, 139.76712]

#10進数から60進数に変換する関数
def convertDeg(coord):
    deg = int(coord)  #度
    min10 = (coord - deg) * 60  #分の10進数表記
    min60 = int(min10)  # 分
    sec = format((min10 - min60) * 60, '.5f')  #秒　小数点以下5桁まで
    return deg, min60, sec  #3つの要素を持つリストが返ってくる
    
#緯度と経度を関数で処理
lat60 = convertDeg(latlon[0])
lon60 = convertDeg(latlon[1])

#結果をprintで出力
print(f"緯度：{lat60[0]}度{lat60[1]}分{lat60[2]}秒")
print(f"経度：{lon60[0]}度{lon60[1]}分{lon60[2]}秒")
```
