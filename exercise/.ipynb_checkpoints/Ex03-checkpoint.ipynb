# 解答案
何をやっているのかひとつずつ追えるようにやや冗長に書いてます。  
とりあえず今は動けばヨシ！

```python
#10進数で表記された緯度経度を60進数に変換して表示する

#東京駅の緯度経度
latlon = [35.68131, 139.76712]

#10進数から60進数に変換する関数
def convertDeg(coord):
    deg = int(coord)  #度
    min10 = abs((coord - deg) * 60)  #分の10進数表記、マイナス（南緯/西経）のときは絶対値absをとる
    min60 = int(min10)  # 分
    sec = round((min10 - min60) * 60, 3)  #秒　小数点以下3桁まで
    return deg, min60, sec  #3つの要素を持つリストが返ってくる
    
#緯度と経度を関数で処理
lat60 = convertDeg(latlon[0])
lon60 = convertDeg(latlon[1])

#結果をprintで出力
print(f"緯度：{lat60[0]}度{lat60[1]}分{lat60[2]}秒")
print(f"経度：{lon60[0]}度{lon60[1]}分{lon60[2]}秒")
```

```python
# リスト内の複数の10進緯度経度を、繰り返し処理によって60進数に変換し、結果を表示する

#緯度経度のリスト
latlons = [
    [35.68131, 139.76712],  #東京
    [78.24885, 15.46454],  #ノルウェー
    [-18.87919, 47.50791],  #マダガスカル
    [38.90719, -77.03687],  #ワシントンD.C.
    [-33.44889, -70.66927]  #チリ
]

for coord in latlons:  #東京、ノルウェー、マダガスカル…と一つずつ取り出してcoordにアサインする
    lat = convertDeg(coord[0])  #緯度
    lon = convertDeg(coord[1])  #経度
    print(f"緯度：{lat[0]}:{lat[1]}:{lat[2]},経度：{lon[0]}:{lon[1]}:{lon[2]}")
```

```python
#数理気候区分に基づき、緯度経度を気候帯に分別する
#緯度経度はlatlonsを使用する

#緯度経度表記を揃える
#ここではせっかくなので1つ前の問題を流用（printではなく新たな変数にアサインする）
#本当は基準の方を10進数に変換したほうが圧倒的に楽
lats60 = []
for coord in latlons:
    lat = convertDeg(coord[0])  #今必要なのは緯度だけ
    lats60.append(lat)

#処理結果を一応確認する
print(lats60)

#条件分岐
#考え方：23度と66度には境目があるので、まず23度と66度のときの分単位の分岐を考える
#23度と66度以外であれば、度単位で分岐させることが可能
for coord in lats60:
    if abs(coord[0]) == 23:  #23度のとき
        if coord[1] < 26:  #26度より小さければ熱帯
            print("熱帯")
        else:  #26度以上なら温帯
            print("温帯")
    elif abs(coord[0]) == 66:  #66度のとき
        if coord[1] > 33:  #33度より大きければ寒帯
            print("寒帯")
        else:  #33度以下なら温帯
            print("温帯")
    elif abs(coord[0]) < 23:  #23度未満なら明らかに熱帯
        print("熱帯")
    elif abs(coord[0]) > 66:  #66度より大きければ明らかに寒帯
        print("寒帯")
    elif abs(coord[0]) > 23 and abs(coord[0]) < 66:  #23度より大きく66度より小さければ明らかに温帯
        print("温帯")
```
